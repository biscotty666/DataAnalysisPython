## Python Time Series Functions

- Decomposing data so the model can understand
- Control the stationary/nonstationary aspect
- Understand the data better

[Video](https://www.youtube.com/watch?v=rBkqkpl21v8&list=PL3N9eeOlCrP5cK0QRQxeJd6GrQvhAtpBK&index=7)

```python
import datetime
import pandas as pd
import requests
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = (10, 8)
mpl.rcParams['axes.grid'] = False
```

```python
coindeskURL = 'https://api.coindesk.com/v1/bpi/historical/close.json?'

start = datetime.date(2017, 1 ,1)
end = datetime.date(2020, 7, 2)

url = f'{coindeskURL}start={start:%Y-%m-%d}&end={end:%Y-%m-%d}'

result = requests.get(url)
result.content
```

**NB. The `decode()` statment was required although it seems to not be necessary in colab**

```python
data = pd.read_json(result.content.decode('utf-8')).iloc[:-2,:1]
data.index.name = 'date'
data.index = pd.to_datetime(data.index)
data
```

```python
data.plot()
```

## Resample

**Resample values**

- D: Calendar day
- B: Business day
- W: Week
- M: Month
- Q: Quarter
- A: Year end
- H: Hours
- M: Minutes
- S: Seconds

```python
data.resample('Q').mean()
```

```python
data.resample('Q').mean().plot()
```

```python
data.plot()
data.resample('M').mean().plot()
```

**The monthly level more closely matches the data than the quarterly**

## diff

Create (weak) stationary data from unstationary data

```python
data[:5]
```

```python
data[:5].diff()
```

```python
data[:5].diff(2)
```

```python
pd.concat([data['bpi'], data['bpi'].diff(2), data['bpi'].diff()],axis=1).plot()
```

We have removed the non-stationary part

## Shift

Re-align values and index

```python
data[:5]
```

```python
data[:5].shift()
```

**`tshift()`**

```python
data[:5].tshift(2)
```

```python
print(data[:5])
print(data[:5] - data[:5].shift())
```

This is the same value returned by `diff()`

```python
data[:5] - data[:5].shift(2)
```

```python
pd.concat([data, data-data.shift(),data.shift(365)],axis=1).plot()
```

```python
data.rolling(window=10).mean().plot()
```

```python
data.resample('W').mean().rolling(window=10).mean().plot()
```

```python
data.rolling(window=50).mean().plot()
data.rolling(window=50, win_type='gaussian').sum(std=10).plot()
```

`data.rolling(window=50, win_type='gaussian').sum(std=10).plot()` is equivalent to Exponential Moving Average (`ewm()`)

```python
data.ewm(span=50).mean().plot()
```

```python
pd.concat([data, data.rolling(window=100).mean(),data.ewm(span=100).mean()],axis=1).plot()
```

### `cumsum()`

### `pct_change()`

```python
data[:5].pct_change(2) # compares to two years prior
```

### `expanding()`

Combined with `sum()` this is the same as `cumsum()` but it can be used with other aggregations

```python
data[:5].expanding().sum() # Same as `cumsum()`
```

```python
data[:5].expanding().mean()
```
