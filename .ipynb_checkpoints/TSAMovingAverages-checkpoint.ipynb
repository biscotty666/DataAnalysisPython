```python
def parse(x):
  return datetime.strptime(x, '%m/%d/%Y')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/electricity_consumption.csv', parse_dates = ['Bill_Date'],date_parser=parse)
df.head()
```

```python
print("Rows   : ", df.shape[0])
print("Columns: ", df.shape[1])
print("\nFeatures:  \n", df.columns.tolist())
print("\nMissing values:    ", df.isnull().any())
print("\nUnique values:    ", df.nunique())
```

```python
bill_df = df.set_index("Bill_Date")
```

```python
bill_df.head()
```

```python
bill_2018 = bill_df['2016':'2018'][['Billed_amount']]
bill_2018
```

The data is not stationary so moving averages will probably not be useful. But let's see...

## Simple moving average


```python
bill_2018['Billed_amount'].rolling(window=3).mean()
```

`shift()` the data in the new column down so prediction aligns with actual

```python
bill_2018['ma_rolling_3']=bill_2018['Billed_amount'].rolling(window=3).mean().shift(1)
bill_2018
```

```python
bill_2018.plot()
```

As expected, rolling average does not perform well on non-stationary data

## Weighted moving average

Weighted moving averages are more sensitive to data points
- Can identify trends sooner than simple moving averages
- Adds complexity because weights must be defined manually

```python
def wma(weights): 
  def calc(x):
    return (weights*x).mean()
  return calc
     
```

```python
bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5])))
```

```python
bill_2018['wma_rolling_3']=bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
```

```python
bill_2018
```

```python
bill_2018.plot()
```

## Exponential Moving Average

[Exponential Moving Average](https://www.educba.com/exponential-moving-average-formula/)

<img src="EMA.jpeg" height='250' width='400'/>

$(Close - previous EMA) * (2 / (span+1)) + previous EMA$

```python
bill_2018['Billed_amount'].ewm(span=3,
                               adjust=False,
                               min_periods=0).mean()
```

```python
bill_2018['ewm_window_3']=bill_2018['Billed_amount'].ewm(span=3,
                               adjust=False,
                               min_periods=0).mean().shift(1)
bill_2018
```

```python
bill_2018.plot()
```

## Exponential Smoothing

<img src="ExpSmoothing.png" />

EMA did not require additional parameters, but Exponential smoothing requires `alpha` for weighting which controls the smoothing decay over time. Larger alphas give more weight to more recent data points.

[Exponential smoothing math](https://byjus.com/maths/exponential-smoothing/)


```python
bill_2018['Billed_amount'].ewm(alpha=0.7,
                               adjust=False,
                               min_periods=3).mean()
```

```python
bill_2018['esm_window_3_7']=bill_2018['Billed_amount'].ewm(alpha=0.7,
                               adjust=False,
                               min_periods=3).mean().shift(1)
bill_2018
```

```python
bill_2018[['Billed_amount','esm_window_3_7']].plot()
```

```python
bill_2018.plot()
```

```python
bill_2018['Billed_amount'].ewm(alpha=0.3,
                               adjust=False,
                               min_periods=3).mean()
```

```python
bill_2018['esm_window_3_3']=bill_2018['Billed_amount'].ewm(alpha=0.3,
                               adjust=False,
                               min_periods=3).mean().shift(1)
bill_2018
```

```python
bill_2018[['Billed_amount','esm_window_3_7','esm_window_3_3']].plot()
```

3_7 is more accurate but 3_3 provides a smoother line capturing the trends

<img src="RMSE.png" />

```python
((bill_2018['Billed_amount']-bill_2018['ma_rolling_3'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['wma_rolling_3'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['ewm_window_3'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['esm_window_3_7'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['esm_window_3_3'])**2).mean()**0.5
```

Exponential Smoothing 3_7 perfomed best. Never-the-less for this type of data more sophisticated models such as Arima or Prophet would perform better.
