```python
pd.concat([df['Appliances'],df['Appliances'].shift(-1),
           df['Appliances'].shift(-2)],axis=1)
```

```python
multi_target=pd.concat([df['Appliances'],
                        df['Appliances'].shift(-1),
                        df['Appliances'].shift(-2)],
                        axis=1).dropna().to_numpy().tolist()
```

**`shift()` will result in `nan` so we need to use `dropna()`**

```python
multi_target[0:13]
```

```python
ts_generator = TimeseriesGenerator(features[:-2],
                                   multi_target,
                                   length=6,
                                   sampling_rate=1,
                                   batch_size=1,
                                   stride=6)
```

By setting length and stride to the same value we create a tumbling/hopping window
